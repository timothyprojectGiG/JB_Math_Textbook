**Section 2.3**

**2.3.1**  $f(z)=[x+iy]^2= x^2-y^2 + i(2xy)$, so $u(x,y)=x^2-y^2$ and $v(x,y)=2xy.$  Thus $u_x=2x  = v_y$  and $u_y=-2y=-v_x$ and the CR-eqns hold with the first partials all continuous functions.  Moreover, we can use CR-eqns to obtain the derivative:  $f(z)= u_x+iv_x=2x+i2y=2z.$ 

**2.3.2**   $f(z)=x^2+y^2$, so $u(x,y)=x^2+y^2$ and $v(x,y)=0.$  Thus $u_x=2x \neq  v_y$  and the CR-eqns do not hold. 

**2.3.3** $f(z)=e^{x+iy}=e^xe^{iy}=e^x\cos y + i e^x  \sin y$.
Thus, $u=e^x \cos y$ and $v=e^x \sin y$. Moreover, $u_x=e^x\cos y = v_y$ and $u_y=-e^x\sin y = -v_x$ so the  C-R eqns hold with $u_x,v_x,u_y,v_y$ all continuous, implying that $f(z)$ is differentiable. Finally, $f'(z)= u_x+iv_x= e^x\cos y + i e^x \sin y = e^z$ as expected.

**2.3.4** Let $f(z)=\log z = \ln r + i\theta$. Then $u=\ln r$ and $v=\theta$. Then $u_r=\frac{1}{r} = \frac{1}{r} v_{\theta}$ and $\frac{1}{r}u_{\theta}= 0 = -v_r$ so $f(z)$ is differentiable (except when $r=0$).  Moreover,

$$
    \frac{d}{dz} \log z = e^{-i\theta}[u_r+iv_r] = e^{-i\theta}\frac{1}{r}=\frac{1}{z},
$$

 so the same formula for real calculus holds for the complex derivative of $\log z$.
